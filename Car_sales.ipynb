##Graphical image to understand easily

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['sellingprice'], bins=20, kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

This show visualize the distribution of selling prices in the dataset

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='odometer', y='sellingprice', data=df)
plt.title('Scatter Plot of Odometer vs. Selling Price')
plt.xlabel('Odometer (Mileage)')
plt.ylabel('Selling Price')
plt.show()
```

Here we can see that the sales of Odometer it is higger at 0.0 and goes upto 160000 and 1.6 odometer has also some selles

#### year and sellingprice


```python
sns.scatterplot(x='year', y='sellingprice', data=df)
```

In this we can that the sale are higger in 2015 as compared to other years and is shows that the sales contiously increasing from 1985 to 2015

###Top 10 Most Common Car Makes

```python
# Bar plot of frequency of car makes
plt.figure(figsize=(10, 8))
sns.countplot(y='make', data=df, order=df['make'].value_counts().index[:10])
plt.title('Top 10 Most Common Car Makes')
plt.xlabel('Frequency')
plt.ylabel('Car Make')
plt.show()
```

From the above graph we conclude that Ford is the number one maker and the series goes like this Chevrolet,Nissan,Toyota,Dodge,BMW,Honda, infiniti,Mercedes-Benz,Hundai in sequence

###Top 10 Most Common Car Model

```python
# Bar plot of frequency of car makes
plt.figure(figsize=(8, 8))
sns.countplot(y='model', data=df, order=df['model'].value_counts().index[:10])
plt.title('Top 10 Most Common Car Model')
plt.xlabel('Frequency')
plt.ylabel('Car Make')
plt.show()
```

Here are top 10 most common car model are 3Series, F-150, G Sedan,
Altima, Accord, Escape, Fusion, Civic, Camry, Impala

###Comparison of Car Models by Sales

```python
model_avg_price = df.groupby('model')['sellingprice'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=model_avg_price.values, y=model_avg_price.index, palette='viridis')
plt.title('Comparison of Car Models by Selling Price')
plt.xlabel('Average Selling Price')
plt.ylabel('Car Model')
plt.show()
```

This graphs shows the top 10 models with the highest average selling prices and according to that Ghost model has higgest average Selling price and in second position is  California and so on

###Top 10 Car Body Types by Selling Price

```python
body_avg_price = df.groupby('body')['sellingprice'].mean().sort_values(ascending=False).head(10)

# Plot the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=body_avg_price.values, y=body_avg_price.index, palette='viridis')
plt.title('Top 10 Car Body Types by Selling Price')
plt.xlabel('Average Selling Price')
plt.ylabel('Car Body Type')
plt.show()
```

Here we can see that Q60Convertible is sold higher as compared to other body type and obove its show Top 10 Car Body Types by Selling Price

```python
# Bar plot comparing car colors by sales
plt.figure(figsize=(10, 6))
sales_by_color = df['color'].value_counts().sort_values(ascending=False)[:10]
sns.barplot(x=sales_by_color.values, y=sales_by_color.index, palette='viridis')
plt.title('Comparison of Car Colors by Sales')
plt.xlabel('Number of Sales')
plt.ylabel('Car Color')
plt.show()
```

The Car which sold higer is most of black in color and the second is White the gray and so

```python
seller_counts = df['seller'].value_counts().head(10)

# Plot the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=seller_counts.values, y=seller_counts.index, palette='viridis')
plt.title('Top 10 Sellers by Number of Cars Sold')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Seller')
plt.show()
```

From this we can conclude that Standard consumer doing the highest sales then other and its almost more than 600 cars and in 2nd nissa infiniti they sold around 500 cars
